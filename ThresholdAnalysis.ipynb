## Precision Recall Analysis

### Average Filter

```python
#Paths
filtername = 'mean_filter'
path = '/jupyter-workspace/private/reco_gui/filter_analysis/%s/conf_matrix/' %(filtername)

f1_plot = []
std_plot = []
for size in range(3,23,2):
    prec_list = []
    recall_list = []
    f1_list = []
    f1_std = []
    for threshold in range(11):
        filename = 'threshold_%.1f_window_%d.npy' %(threshold, size)

        conf_matrix = np.load(path+filename)

        precision = conf_matrix[0:200,1,1] / (conf_matrix[0:200,0,1] + conf_matrix[0:200,1,1])
        recall = conf_matrix[0:200,1,1] / (conf_matrix[0:200,1,0] + conf_matrix[0:200,1,1])

        if sum(np.isnan(precision) != 0):
            precision[np.where(np.isnan(precision))] = 0
            f1_score[np.where(np.isnan(precision))] = 0
        else:
            f1_score = 2 * (precision * recall) / (precision + recall)

        prec_list.append(np.mean(precision))
        recall_list.append(np.mean(recall))
        f1_list.append(np.mean(f1_score))
        f1_std.append(np.std(f1_score))

    print('Maximum f1 score of %s - window size %d is %.5f +- %.5f with threshold equal to %.1f' %(filtername, size, np.max(f1_list), f1_std[np.argmax(f1_list)], np.argmax(f1_list)))
    f1_plot.append(np.max(f1_list))
    std_plot.append(f1_std[np.argmax(f1_list)])
```

```python
font = 20

fig, ax = plt.subplots(figsize=[10,7])

window_size = np.arange(3,23,2)

ci = 1.96 * np.array(std_plot) / np.sqrt(200)

ax.fill_between(window_size, (np.array(f1_plot) + ci), (np.array(f1_plot) - ci) , color='b', alpha=.2)

ax.set_title('Average Filter F1 scores', fontsize=font)
ax.set_xlabel('Window size', fontsize=font)
ax.set_ylabel('F1 score', fontsize=font)
ax.set_xticks(np.arange(1,23,2))
ax.grid()
```

### Cygno pre processing

```python
runpath = '/jupyter-workspace/private/reco_gui/runs_fe/'
recopath = '/jupyter-workspace/private/reco_gui/reco_files/simulation_6kev/'

filename = 'reco_run20001_3D_cygno_p1_r4.root'

f = ROOT.TFile(recopath + filename)
tree = f.Get('Events')

integral = root_numpy.tree2array(tree, branches = 'sc_integral')
x_pix = root_numpy.tree2array(tree, branches = 'redpix_ix')
y_pix = root_numpy.tree2array(tree, branches = 'redpix_iy')

integral_truth = []

conf_matrix_list = []

for ev in range(200,300):
    run = (ev // 30) + 1
    runev = ev % 30
    
    f = ROOT.TFile(runpath + 'histograms_Run%05d.root' %(run))
    img_fe = root_numpy.hist2array(f.Get('pic_run%d_ev%d' %(run, runev)))[128:-128, 128:-128]
    f.Close()
    
    xx = x_pix[ev]
    yy = y_pix[ev]
    
    img_clu = np.zeros([2048,2048])
    img_clu[xx,yy] = 1
    
    conf_matrix = metrics.confusion_matrix((img_fe > 0).flatten(), (img_clu > 0).flatten())
    
    conf_matrix_list.append(conf_matrix)
    
    integral_truth.append(sum(sum(img_fe)))
    print('Event %d done.' %(ev))

cm_arr = np.asarray(conf_matrix_list)
int_truth = np.asarray(integral_truth)

precision = cm_arr[0:100,1,1] / (cm_arr[0:100,0,1] + cm_arr[0:100,1,1])
recall = cm_arr[0:100,1,1] / (cm_arr[0:100,1,0] + cm_arr[0:100,1,1])

f1_score = 2 * (precision * recall) / (precision + recall)

```

```python
print('Cygno preprocessing: Precision = %.4f \u00B1 %.4f and Recall = %.4f \u00B1 %.4f' %(np.mean(precision), np.std(precision)/np.sqrt(100), np.mean(recall), np.std(recall)/np.sqrt(100)))
print('Cygno preprocessing: F1_score = %.4f \u00B1 %.4f' %(np.mean(f1_score), np.std(f1_score)/np.sqrt(100)))
```
